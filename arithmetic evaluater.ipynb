```python
#Assumption : don't use alphabets and use only digits less than 10 and don't divide by 0 and don't use power operation. 
# dont use "[]" or "{ }" bracket ,use only "( )" bracket. symbols:multiplication (*), division(/),add(+),subraction(-)

inp = input("enter your expression")
operand = []
stack=[]
def prec(x):
    if((x=="(") | (x==")")):
        return 3
    elif((x=='*') | (x=="/")):
        return 2
    elif((x=="+") | (x=="-")):
        return 1

def opp(a,b,op):
    if(op=="*"):
        return a*b
    elif(op=="/"):
        return a/b
    elif(op=="+"):
        return a+b
    elif(op=="-"):
        return a-b
for  x in inp:
    if( x in '0123456789'):
        operand.append(x)
    
    elif(x=='('):
        stack.append(x)
        
    elif(x==')'):
        y=-1
        while(stack[y]!='('):
            operand.append(stack.pop())
           
        stack.pop()
        
    elif(x=='*'):
        if(len(stack)==0):
            stack.append(x)
        elif(prec(stack[-1])==2):
            operand.append(stack.pop())
            stack.append(x)
        else:
            stack.append(x)
            
    elif(x=='/'):
        if(len(stack)==0):
            stack.append(x)
        elif(prec(stack[-1])==2):
                operand.append(stack.pop())
                stack.append(x)
        else:
            stack.append(x)
    elif(x == '+'):
        if(len(stack)==0):
            stack.append(x)
        elif(prec(stack[-1])==3):
            stack.append(x)
        elif(len(stack)!=0):
            if(prec(stack[-1])==1):
                operand.append(stack.pop())   
                stack.append(x)
            elif(prec(stack[-1])==2):
                while(prec(stack[-1])==2):
                    operand.append(stack.pop())
                    if(len(stack)==0):
                        break
                stack.append(x)
    elif(x=="-"):
        if(len(stack)==0):
            stack.append(x)
        elif(stack[-1]=='-'):
            operand.append(stack.pop())
            stack.append(x)
        elif(stack[-1]=="+"):   
            operand.append(stack.pop())       
            stack.append(x)
        elif(len(stack)!=0):
            while(stack[-1]!="("):
                operand.append(stack.pop())
                if(len(stack)==0):
                    break
            stack.append(x)
    else:
        pass
while(len(stack)!=0):
    operand.append(stack.pop())
    
for x in operand:
    if(x in '0123456789'):
        stack.append(x)
    elif(x == '+'):
        a=float(stack.pop())
        b=float(stack.pop())
        stack.append(opp(b,a,x))
    elif(x == '-'):
        a=float(stack.pop())
        b=float(stack.pop())
        stack.append(opp(b,a,x))
    elif(x == '*'):
        a=float(stack.pop())
        b=float(stack.pop())
        stack.append(opp(b,a,x))
    elif(x == '/'):
        a=float(stack.pop())
        b=float(stack.pop())
        stack.append(opp(b,a,x))

print("The evaluated value is:")   
print(stack)

print(operand)
```
